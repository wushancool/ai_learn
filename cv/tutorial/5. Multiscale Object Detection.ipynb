## Multiscale Object Detection
Reduce the number of anchor boxes.
1. Uniform sampling
2. The number of smaller anchor box is more than bigger anchor box.

```python
from img import img_withbox
from cv.utils.anchorbox import *
from PIL import Image
from torchvision.transforms import ToTensor

img = "cv/tutorial/assets/catdog.jpg"
to_tensor = ToTensor()
h, w= to_tensor(Image.open(img)).shape[-2:]
scale = torch.tensor((w,h,w,h))

def show_anchors(h, w, s):
    boxes = multibox_prior((h,w), s, [1, 2, 0.5])
    img_withbox(img, boxes * scale)

show_anchors(4,4, [0.2])
```

```python
show_anchors(2,2, [0.4])
```

```python
show_anchors(1,1,[0.7])
```
